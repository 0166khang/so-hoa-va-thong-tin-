```python
## cho 1 bản np có 150 dòng 5 cột
import numpy as np
from sklearn.model_selection import train_test_split

# 1. Tạo mảng NumPy 150 x 5
np.random.seed(42)  # Đặt seed để kết quả nhất quán
data = np.random.rand(150, 5) * 100  # Tạo mảng ngẫu nhiên, nhân với 100 để tăng giá trị

# 2. Chia cột đầu tiên (chiều cao) thành x và cột cuối cùng (GPA) thành y
x = data[:, 0]  # Cột đầu tiên
y = data[:, -1]  # Cột cuối cùng

# 3. Chia dữ liệu x và y thành train/test (70:30)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

# 4. Tạo tập dữ liệu không chồng chéo (loại bỏ giá trị trùng lặp trong x_train)
x_train_unique = np.unique(x_train)


```

```python
## bài tập về nhà tết
def print_board(board):
    for row in board:
        print(" ".join(str(cell) for cell in row))
    print()

def check_winner(board, player):
    for row in board:
        if all(cell == player for cell in row):
            return True

    for col in range(3):
        if all(board[row][col] == player for row in range(3)):
            return True

    if all(board[i][i] == player for i in range(3)) or all(board[i][2 - i] == player for i in range(3)):
        return True

    return False

def tic_tac_toe():
    board = [[99 for _ in range(3)] for _ in range(3)]
    players = {0: 'O', 1: 'X'}
    turn = 1  # X đi trước

    print("Bài tập tết về nhà - Trò chơi cờ caro")
    print_board(board)

    for _ in range(9):
        while True:
            try:
                row, col = map(int, input(f"Người chơi {players[turn]}, nhập vị trí (hàng, cột): ").split())
                if 0 <= row < 3 and 0 <= col < 3 and board[row][col] == 99:
                    board[row][col] = players[turn]
                    break
                else:
                    print("Vị trí không hợp lệ hoặc đã có người chọn. Hãy thử lại!")
            except ValueError:
                print("Vui lòng nhập hai số cách nhau bằng dấu cách!")

        print_board(board)

        if check_winner(board, players[turn]):
            print(f"Người chơi {players[turn]} thắng!")
            return

        turn = 1 - turn  # Đổi lượt chơi

    print("Hòa! Trò chơi kết thúc.")

tic_tac_toe()
```
